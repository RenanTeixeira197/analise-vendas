```python
df = pd.read_csv('train.csv')
tem_negativos = df[df['Sales'] < 0]
print(f"Valores Negativos: {tem_negativos}")
```

Dados Incoerentes (Ship Date antes de Order Date)

```python
dados_incoerentes = df[df['Ship Date'] < df['Order Date']]
print(dados_incoerentes)
#
```

Corrigindo datas usando dayfirst=True

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

dados_incoerentes_apos_correcao = df[df['Ship Date'] < df['Order Date']]
print("Dados incoerentes após correção:")
print(dados_incoerentes_apos_correcao)
```

```python
dados_incoerentes = df[df['Ship Date'] < df['Order Date']]
print(dados_incoerentes)
```

Criando Colunas (Ano, Mês e Tempo de Entrega)

```python
ano = df['Order Date'].dt.year
mes = df['Order Date'].dt.month
tempo_entrega = df['Ship Date'] - df['Order Date']
```

Receita Total: A receita total no período analisado foi de $2.261.536,79

```python
receita_total = df['Sales'].sum()
print(f"Receita Total: ${receita_total}")
```

Número de Pedidos: Quantos Pedidos foram realizados?

```python
numero_de_pedidos = df['Order ID'].nunique()
print(f"Número de Pedidos: {numero_de_pedidos}")
```

Ticket Médio $459.48

```python
ticket_medio = receita_total / numero_de_pedidos
print(f"Ticket Médio: ${ticket_medio}")
#
```

BLOCO 2 - MÉTRICAS TEMPORAIS

Receita por Ano

```python
receita_por_ano = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
print(receita_por_ano)
#
```

Receita por Mês (Análise de Sazonalidade)

```python
receita_por_mes = df.groupby(df['Order Date'].dt.month)['Sales'].sum()
print(receita_por_mes)
```

BLOCO 3: PERFORMANCE POR LOCALIZAÇÃO

Receita por Região

```python
receita_por_regiao = df.groupby('Region')['Sales'].sum()
print(receita_por_regiao)
#
```

Receita por Estado

```python
receita_por_estado = df.groupby('State')['Sales'].sum()
print(receita_por_estado)
#
```

Bloco 4: Produtos

Receita por Categoria

```python
receita_por_categoria = df.groupby('Category')['Sales'].sum()
print(receita_por_categoria)
#
```

Top 10 Produtos por Receita

```python
produtos_mais_lucrativos = df.groupby('Product Name')['Sales'].sum().nlargest(10)
print(produtos_mais_lucrativos)
#
```

Receita por Subcategoria

```python
subcategorias_mais_lucrativas = df.groupby('Sub-Category')['Sales'].sum().nlargest(10)
print(subcategorias_mais_lucrativas)
#
```

Bloco 5: Seguimento de Cliente

Receita por Segmento

```python
receita_por_seguimento = df.groupby('Segment')['Sales'].sum()
print(receita_por_seguimento)
#
```

Receita por Ano:

```python
receita_por_ano.pct_change() * 100
```
